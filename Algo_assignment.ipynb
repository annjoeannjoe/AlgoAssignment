```python
#Algorithm

#1st Algorithm: KMP string matching algorithm
#preprosing to calculate LPS
def computeLPSArray(pat, M, lps):
    len = 0 
    lps[0] 
    i = 1
  
    while i < M:
        if pat[i]== pat[len]:
            len += 1
            lps[i] = len
            i += 1
        else:
            if len != 0:
                len = lps[len-1]
            else:
                lps[i] = 0
                i += 1

def KMPSearch(pat, txt):
    M = len(pat)
    N = len(txt)

    if(M!=N):
      return False

    lps = [0]*M
    j = 0 
  
    computeLPSArray(pat, M, lps)
    isFound = False
    i = 0 

    while i < N:
        if pat[j] == txt[i]:
            i += 1
            j += 1

        if j == M:
            j = lps[j-1]
            isFound = True
        elif i < N and pat[j] != txt[i]:
            if j != 0:
                j = lps[j-1]
            else:
                i += 1
  
    return isFound



#2nd Algorithm: Reverse combsort algorithm    
#Proprocessing: This function used to find next gap from current
def getNextGap(gap):
    gap = (gap * 10)/13
    if gap < 1:
        return 1
    return gap
 
def reverse_combSort(arr):
    n = len(arr)
    gap = n
    swapped = True

    while gap !=1 or swapped == 1:
        gap = int(getNextGap(gap))
        swapped = False

        for i in range(0, n-gap):
            if arr[i] < arr[i + gap]:
                arr[i], arr[i + gap]=arr[i + gap], arr[i]
                swapped = True
                
    return arr
```

```python
import string
import plotly.express as px
import pandas as pd
import nltk
import plotly.io as pio
import plotly.graph_objects as go
#Tokenizer
nltk.download('punkt')
#StopWords Library
nltk.download('stopwords')
from nltk.corpus import stopwords

#1st method
#
def getTextList(file):
  f= open(f"{file}.txt","r")
  textList = f.read().split(",")
  textList = [text.strip().lower() for text in textList]

  return textList


#2nd method
#Clean all the non usable characters such as punctuation, single-character digit, empty string in the article text list
def cleanNoise(articleText_list):
  #make all the word in the articleText_list to lowercase
  articleText = " ".join(articleText_list)
  articleText = articleText.lower()
  
  #remove all the punctuations in the article text
  translatable =str.maketrans("" , "", string.punctuation) 
  articleText = articleText.translate(translatable)  

  #Split the article text that without noise into list element back
  articleTextWithoutNoise_list = articleText.split(" ")
  
  #eliminate empty string & digits & single alphabet character
  articleTextWithoutNoise_list = [substring.lower() for substring in articleText_list if substring != "" and not substring.isdigit() and len(substring)>1]

  #return list that without noise
  return articleTextWithoutNoise_list



#3rd method
#Remove stopword using built-in stopword from nltk library. The comparison is limited to English and done by using KMP algorithm.
def cleanStopword(articleText_list):
  stopword_list = stopwords.words('english')

  for s in stopword_list:
    for t in articleText_list:
      if KMPSearch(s,t):
        articleText_list.remove(t)

  articleTextWithoutStopword_list = articleText_list
  return articleTextWithoutStopword_list



#4nd method
#Calculate the occurence of each word in the article text. The return formt of the tuple list is: 6, Hello ......
def wordCount_tupList(articleText_list):
  #Remove repeated word in the article text list
  wordList = list(set(articleText_list))
  return [(articleText_list.count(a) , a) for a in wordList] 



#5nd method**********************************************************
#
def plotBarGraph(wordCount_tup_lst , article_title, bar_keyword, outputName):
  #zip the tuple_list
  if len(wordCount_tup_lst) == 0:
    count, word = [], []
  else:
    count, word = zip(*wordCount_tup_lst)

    #Syntax px.bar(x-axis data (a tuple),y-axis data (a tuple),labels (pass in a dictionary),title,template,height,color_discrete_sequence =['green']*3)

  title="Bar Chart of "+ bar_keyword +" for Article [ "+ article_title+" ]"
  labels={'x':'Word','y':'Count'}

    #dict(x='Word', y = 'Count')
    #Templates from plotly.io
    #pio.templates
    #['ggplot2', 'seaborn', 'simple_white', 'plotly','plotly_white', 'plotly_dark', 'presentation', 'xgridoff','ygridoff', 'gridon', 'none']

  fig = px.bar(x= word, y=count, labels = labels ,title=title,template='seaborn')
  
  # Change color of bar graph
  # ,color_discrete_sequence =['green']*3
  #fig.update_traces(marker_color='yellow')
  
  #,text=count
  #fig.update_traces(textfont_size=20, textangle=0, textposition="auto", cliponaxis=False)

  # ADD TEXT ON TOP OF BAR GRAPH
  #,text_auto='.2s'
  #fig.update_traces(textfont_size=20, textangle=0, textposition="outside", cliponaxis=False)
  
  fig.show()

#6th method to plot pie chart
def PlotPieChartAnalysis(count,country_name,neutralWordCount,positiveWordCount,negativeWordCount):
  #,pull=[0, 0.2, 0.3]
  c= count+1
  s=str(c)
  title = "Neutral Word & Positive Word & Negative Word Count Analysis for Article " + s +" which related to "+ country_name +" :"
  colors = ['royalblue', 'darkblue', 'lightcyan']

  fig = go.Figure(data=[go.Pie(labels=['Neutral Word Count','Positive Word Count','Negative Word Count'],
                             values=[neutralWordCount,positiveWordCount,negativeWordCount])])
  fig.update_traces(hoverinfo='label+value', textinfo='value+percent', textfont_size=12,
                  marker=dict(colors=colors, line=dict(color='#000000', width=2)))
  fig.update_layout(title_text=title, title_x=0.4)
  fig.show()

#7th method
#Check the occurence of word in particular word list by using kmp string matching algorithm
def exist(word, word_list):
  for item in word_list:
    if KMPSearch(item,word):
      return True
  return False



#8th method
#clssify the type of word (positive, nagataive, neutral) by using kmp string matching algorithm
def classifyWordType(positiveWord_list, negativeWord_list,articleText_list):
  positiveWordinText_list , negativeWordinText_list, neutralWordinText_list = [], [], []

  for word in articleText_list:
    if exist(word,positiveWord_list):
      positiveWordinText_list.append(word)
    elif exist(word,negativeWord_list):
      negativeWordinText_list.append(word)
    else:
      neutralWordinText_list.append(word)     

  return positiveWordinText_list, negativeWordinText_list, neutralWordinText_list



#9th method
#Get score of each word type by dividing the occurance with total word return 3 float values
def getSentimentScore(positiveList, negativeList, neutralList ,totalList):
  if len(positiveList) ==0:
    positiveScore = 0
    negativeScore = len(negativeList)/len(totalList)
  elif len(negativeList) ==0:
    negativeScore = 0
    positiveScore = len(positiveList)/len(totalList)
  else :
    positiveScore = len(positiveList)/len(totalList)
    negativeScore = len(negativeList)/len(totalList)

  neutralScore = len(neutralList)/len(totalList)

  return positiveScore, negativeScore, neutralScore



```

```python
def calcMinRun(n):
  
    MIN_MERGE = 16
    r = 0
    while n >= MIN_MERGE:
        r |= n & 1
        n >>= 1
    return n + r


# This function sorts array from left index to right index which is of size atmost RUN
def insertionSort(arr, left, right):
    for i in range(left + 1, right + 1):
        j = i
        while j > left and arr[j] < arr[j - 1]:
            arr[j], arr[j - 1] = arr[j - 1], arr[j]
            j -= 1


# Merge function merges the sorted runs
def merge(arr, l, m, r):
    # original array is broken in two parts
    # left and right array
    len1, len2 = m - l + 1, r - m
    left, right = [], []
    for i in range(0, len1):
        left.append(arr[l + i])
    for i in range(0, len2):
        right.append(arr[m + 1 + i])

    i, j, k = 0, 0, l

    # after comparing, we merge those two array in larger sub array
    while i < len1 and j < len2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1

        else:
            arr[k] = right[j]
            j += 1

        k += 1

    # Copy remaining elements of left, if any
    while i < len1:
        arr[k] = left[i]
        k += 1
        i += 1

    # Copy remaining element of right, if any
    while j < len2:
        arr[k] = right[j]
        k += 1
        j += 1


# Iterative Timsort function to sort the array[0...n-1] (similar to merge sort)
def timSort(arr):
    n = len(arr)
    minRun = calcMinRun(n)

    # Sort individual subarrays of size RUN
    for start in range(0, n, minRun):
        end = min(start + minRun - 1, n - 1)
        insertionSort(arr, start, end)

    # Start merging from size RUN (or 32). It will merge to form size 64, then 128, 256 and so on ....
    size = minRun
    while size < n:

        # Pick starting point of left sub array. We
        # are going to merge arr[left..left+size-1]
        # and arr[left+size, left+2*size-1]
        # After every merge, we increase left by 2*size
        for left in range(0, n, 2 * size):

            # Find ending point of left sub array
            # mid+1 is starting point of right sub array
            mid = min(n - 1, left + size - 1)
            right = min((left + 2 * size - 1), (n - 1))

            # Merge sub array arr[left.....mid] &
            # arr[mid+1....right]
            if mid < right:
                merge(arr, left, mid, right)

    size = 2 * size

def reverse_timSort(arr):
    n = len(arr)
    gap = n
    swapped = True

    while gap !=1 or swapped == 1:
        gap = int(getNextGap(gap))
        swapped = False

        for i in range(0, n-gap):
            if arr[i] < arr[i + gap]:
                arr[i], arr[i + gap]=arr[i + gap], arr[i]
                swapped = True
                
    return arr

probabilityArray=[]
def calcProbability(problem1_list, problem2_list):
  sum1=0
  sum2=0
  x=0
  while(x<len(problem1_list)):
      sum1 += problem1_list[x]
      sum2 += problem2_list[x]
      x+=1

  countryName=["MX","CN","US","JP","CA"]
  y=0
  while (y < len(problem1_list)):
      probability = (problem1_list[y]/sum1 )* (1-(problem2_list[y]/sum2))
      probabilityArray.append((probability,countryName[y]))
      y+=1

  timSort(probabilityArray)
  reverse_timSort(probabilityArray)
  return probabilityArray

# Print result
calcProbability(finalscore,distanceList)

print("         Ranking          |         Country          |         Probability          ")
print("________________________________________________________________________________")
x=0
while x<len(probabilityArray):
    print("           ",x+1,"            |          ",probabilityArray[x][1],"            | "+"         ",round(probabilityArray[x][0],5),"   ")
    print("________________________________________________________________________________")
    x+=1
```
